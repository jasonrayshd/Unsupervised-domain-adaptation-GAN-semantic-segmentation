```python
dest = "/home/admin/segmentation/task2/datasets"
dataset_name1 = "potsdam"
dataset_name2 = "vaihingen"


import matplotlib.pyplot as plt
from PIL import Image
import os
import re
```

```python
training_files_images1 = sorted(os.listdir(f"{dest}/{dataset_name1}/train/images"))
training_files_labels1 = sorted(os.listdir(f"{dest}/{dataset_name1}/train/labels"))
training_files_images1
training_files_labels1
```

```python
import torch
from torchvision import transforms
```

```python
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.ToPILImage(),
])
```

```python
size = 512
```

## split dataset potsdam images and labels

```python
for index, data in enumerate(training_files_images1):
    img1 = Image.open(f"{dest}/{dataset_name1}/train/images/{training_files_images1[index]}")
    img2 = Image.open(f"{dest}/{dataset_name1}/train/labels/{training_files_labels1[index]}")
    name = re.findall("\d_\d",data)[0]
    img1 = transform(img1)
    img2 = transform(img2)
    for i in range(11):
#         left, upper, right, and lower
        for j in range(11):
            print(f"now position :{(i*size,j*size,(i+1)*size,(j+1)*size)}")
            raw1 = img1.crop((i*size,j*size,(i+1)*size,(j+1)*size))
            raw2 = img2.crop((i*size,j*size,(i+1)*size,(j+1)*size))
            raw1.save(f"{dest}/{dataset_name1}/train/cropped/images/{name}_{i}_{j}.png")
            raw2.save(f"{dest}/{dataset_name1}/train/cropped/labels/{name}_{i}_{j}.png")
```

```python
training_files_images2 = sorted(os.listdir(f"{dest}/{dataset_name2}/train/images"))
training_files_labels2 = sorted(os.listdir(f"{dest}/{dataset_name2}/train/labels"))
print(training_files_images2)
print(training_files_labels2)
```

## split dataset vaihingen images and labels

```python
for index, data in enumerate(training_files_images1):
    img1 = Image.open(f"{dest}/{dataset_name2}/train/images/{training_files_images2[index]}")
    img2 = Image.open(f"{dest}/{dataset_name2}/train/labels/{training_files_labels2[index]}")
    size = img1.size
    name = re.findall("\d+",data)[0]
    img1 = transform(img1)
    img2 = transform(img2)
    for i in range(size[0] // 512):
#         left, upper, right, and lower
        for j in range(size[1] // 512):
            print(f"now position :{(i*size,j*size,(i+1)*size,(j+1)*size)}")
            raw1 = img1.crop((i*size,j*size,(i+1)*size,(j+1)*size))
            raw2 = img2.crop((i*size,j*size,(i+1)*size,(j+1)*size))
            raw1.save(f"{dest}/{dataset_name2}/train/cropped/images/{name}_{i}_{j}.png")
            raw2.save(f"{dest}/{dataset_name2}/train/cropped/labels/{name}_{i}_{j}.png")
```
