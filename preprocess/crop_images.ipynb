```python
import matplotlib.pyplot as plt
from PIL import Image
import os
import re

import torch
from torchvision import transforms
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.ToPILImage(),
])
size = 512

dest = "/home/admin/segmentation/task2/data"
```

## split training images into train and validation

```python

def split_dataset(training_files_images, training_files_labels, dataset_name, dest, pattern ):
    for index, data in enumerate(training_files_images):
        img1 = Image.open(f"{dest}/{dataset_name}/train/images/{training_files_images[index]}")
        img2 = Image.open(f"{dest}/{dataset_name}/train/labels/{training_files_labels[index]}")
        name = re.findall(pattern,data)[0]
        img1 = transform(img1)
        img2 = transform(img2)
        for i in range(11):
    #         left, upper, right, and lower
            for j in range(11):
                print(f"now position :{(i*size,j*size,(i+1)*size,(j+1)*size)}")
                raw1 = img1.crop((i*size,j*size,(i+1)*size,(j+1)*size))
                raw2 = img2.crop((i*size,j*size,(i+1)*size,(j+1)*size))
                raw1.save(f"{dest}/{dataset_name}/train/cropped/images/{name}_{i}_{j}.png")
                raw2.save(f"{dest}/{dataset_name}/train/cropped/labels/{name}_{i}_{j}.png")
```

## crop  images

```python

def crop_images(image_files,data_path, dest):
    global size
    count = 0
    for index, data in enumerate(image_files):
        img = Image.open(f"{data_path}/{data}")
        img_size = img.size
        img = transform(img)
        for i in range(img_size[0] // 512):
            for j in range(img_size[1] // 512):
                print(f"now position :{(i*size,j*size,(i+1)*size,(j+1)*size)}, image:img{count}_{i}_{j}.png")
                raw1 = img.crop((i*size,j*size,(i+1)*size,(j+1)*size))
               
                raw1.save(f"{dest}/img{count}_{i}_{j}.png")
                count += 1
```

```python
data_path = "/home/admin/segmentation/task2/data/vaihingen/test/images"
dest = "/home/admin/segmentation/task2/data/vaihingen/test/cropped"

image_files = [f for f in os.listdir(data_path) if os.path.isfile(f"{data_path}/{f}")]
crop_images(image_files, data_path, dest)
```
