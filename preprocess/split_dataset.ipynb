## split train and validation dataset in cropped folder

```python
import os

potsdam_root = "/home/admin/segmentation/task2/data/potsdam/train/cropped" # 1452 number of images
vaihingen_root = "/home/admin/segmentation/task2/data/vaihingen/train/cropped" # 210 number of images

# train:val ~= 3:1

def split_train_val(path):
    images = sorted([f for f in os.listdir(f"{path}/images") if os.path.isfile(os.path.join(f"{path}/images/",f))])
    masks = sorted([f for f in os.listdir(f"{path}/masks") if os.path.isfile(os.path.join(f"{path}/masks/",f))])

    assert len(images) == len(masks)
    num = 3*len(images) // 4

    for index, image in enumerate(images):
        if index < num:
            os.rename(f"{path}/images/{image}",f"{path}/images/train/{image}")
        else:
            os.rename(f"{path}/images/{image}", f"{path}/images/val/{image}")

    for index, image in enumerate(masks):
        if index < num:
            os.rename(f"{path}/masks/{image}", f"{path}/masks/train/{image}")
        else:
            os.rename(f"{path}/masks/{image}", f"{path}/masks/val/{image}")
        


split_train_val(potsdam_root)
# split_train_val(vaihingen_root)
```
