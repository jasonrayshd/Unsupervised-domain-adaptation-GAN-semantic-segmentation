```python
dest = "/home/admin/segmentation/task2/datasets"
dataset_name1 = "potsdam"
dataset_name2 = "vaihingen"


import matplotlib.pyplot as plt
from PIL import Image
import os
import re
```

```python
training_files_images1 = os.listdir(f"{dest}/{dataset_name1}/train/images")
training_files_labels1 = os.listdir(f"{dest}/{dataset_name1}/train/labels")
training_files_images1
```

```python
import torch
from torchvision import transforms
```

```python
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.ToPILImage(),
])
```

```python
size = 512
```

## split dataset potsdam

```python
for index, data in enumerate(training_files_images1):
    img = Image.open(f"{dest}/{dataset_name1}/train/images/{data}")
    count = 0
    name = re.findall("\d_\d",data)[0]
    img = transform(img)
    
    for i in range(11):
#         left, upper, right, and lower
        for j in range(11):
            print(f"now position :{(i*size,j*size,(i+1)*size,(j+1)*size)}")
            raw = img.crop((i*size,j*size,(i+1)*size,(j+1)*size))
            raw.save(f"{dest}/{dataset_name1}/train/cropped/images/{name}_{i}_{j}.png")
```
