```python
for index, item in enumerate(train_list1):
    train_list1[index] = item.replace("label","RGB")
#     print(item)
```

```python
count = 0
for index, data in enumerate(dataset1):
    if data in train_list1:
        copyfile(f"{dataset_path1}/{data}",f"{dest}/{dataset_name1}/train/images/{data}")
        count+=1
    elif not "tfw" in data:
        copyfile(f"{dataset_path1}/{data}",f"{dest}/{dataset_name1}/test/{data}")
print(f"copy {count} files finished!")
```

```python
for index, data in enumerate(train_gt_list1):
    copyfile(f"{dataset_gt1}/{data}",f"{dest}/{dataset_name1}/train/labels/{data}")
```

```python
count = 0
for index, data in enumerate(dataset2):
    if data in train_list2:
        copyfile(f"{dataset_path2}/{data}",f"{dest}/{dataset_name2}/train/images/{data}")
        count += 1
    elif not "tfw" in data:
        copyfile(f"{dataset_path2}/{data}",f"{dest}/{dataset_name2}/test/{data}")
        
print(f"copy {count} files finished!")
```
